```python
#import library
import pandas as pd
```

```python

dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python

# here data have categorical value 
# need to change state col value into numerical using one hot encoding

# dataset=pd.get_dummies(dataset)
# dataset
dataset=pd.get_dummies(dataset,drop_first=True) # to reduce the repetative info need to use drop 

dataset=dataset.astype(int) # to change true or flase to int
dataset
```

```python
# to get column names from table
dataset.columns 
```

```python
# need to assign input and output
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
print(independent)
dependent=dataset[['Profit']]
print(dependent)
```

```python
# spliting training and testing data

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
X_train,X_test,Y_train,Y_test
```

```python
#Model creation
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
weight= regressor.coef_
print("Weight:",weight)
bias= regressor.intercept_
print("Bias:",bias)
```

```python
#using test data
test_pred = regressor.predict(X_test)
test_pred
```

```python
Y_test
```

```python
residuals = Y_test - test_pred # to actual output - predicted output
```

```python
import matplotlib.pyplot as plt
plt.scatter(test_pred, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values')
plt.show()
```
