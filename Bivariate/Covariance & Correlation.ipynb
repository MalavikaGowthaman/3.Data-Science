```python
import matplotlib.pyplot as plt  
plt.figure(figsize=(10, 5))
plt.scatter(dataset['degree_p'], dataset['etest_p'], alpha=0.7)
plt.title('Scatter Plot of degree Percentage vs. entrance test Percentage')
plt.xlabel('degree Percentage')
plt.ylabel('entrance Percentage')
plt.grid(True)
plt.show()
```
