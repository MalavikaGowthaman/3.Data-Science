```python
import matplotlib.pyplot as plt  
import seaborn as sns
plt.figure(figsize=(10, 5))
plt.scatter(dataset['degree_p'], dataset['etest_p'])
plt.title("Covariance of degree pass and entrance test pass")
plt.xlabel("degree")
plt.ylabel("entrance")
plt.show()
```
