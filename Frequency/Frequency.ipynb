```python
# to create a table
freqtable = pd.DataFrame(columns =["Unique_values","Frequancy","Relative_Freq","CumulativeFreq"])
freqtable
```

```python
# to get a unique value
freqtable["Unique_values"]=dataset["ssc_p"].value_counts().index
freqtable
```

```python
# to get a counts of value
freqtable["Frequancy"]=dataset["ssc_p"].value_counts().values
freqtable
```

```python
# relative freq = no of occurance/ total no of values
n=len(freqtable)
n
```

```python
freqtable["Relative_Freq"]=dataset["ssc_p"].value_counts().values/n
freqtable
```

```python
# cummulative function --> cumsum (inbuild func)
freqtable["CumulativeFreq"]=freqtable["Relative_Freq"].cumsum() #inbuild func to find cummulative freq
freqtable
```

```python
def freqtable(columnname,dataset):
    freqtable = pd.DataFrame(columns =["Unique_values","Frequancy","Relative_Freq","CumulativeFreq"])
    freqtable["Unique_values"]=dataset[columnname].value_counts().index
    freqtable["Frequancy"]=dataset[columnname].value_counts().values
    n=len(freqtable)
    freqtable["Relative_Freq"]=dataset[columnname].value_counts().values/n
    freqtable["CumulativeFreq"]=freqtable["Relative_Freq"].cumsum() #inbuild func to find cummulative freq
    return freqtable
freqtable("hsc_p",dataset)
```
