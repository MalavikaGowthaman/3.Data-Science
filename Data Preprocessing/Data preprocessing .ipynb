```python
import pandas as pd

# Read data from CSV into a DataFrame
df = pd.read_csv("kidney_disease.csv")
df
```

```python
qual=[]
quan=[]

for columnname in df.columns:
    if (df[columnname].dtypes=='O'):
        qual.append(columnname)
    else:
        quan.append(columnname)
```

```python
qual
```

```python
df.isnull().sum()
```

```python
quan
```

```python
df["rbc"].dtypes
```

```python
df.columns
```

```python
# List of columns you want to fill null values in
columns_to_fill = ['age', 'bp', 'sg']

# Fill null values in specified columns with their respective means
df.fillna(df[columns_to_fill].mean(), inplace=True)
df
```

```python
df.isnull().sum()
```

```python
for col_to_fill in quan:
    mean_value = df[col_to_fill].mean()
    df[col_to_fill].fillna(mean_value, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
for col_to_fill in qual:
    mode_value = df[col_to_fill].mode()[0]
    df[col_to_fill].fillna(mode_value, inplace=True)
df
```

```python
df.isnull().sum()
```
